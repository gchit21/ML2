```python
transaction = pd.read_csv('/kaggle/input/ieee-fraud-detection/test_transaction.csv')
identity = pd.read_csv('/kaggle/input/ieee-fraud-detection/test_identity.csv')
df = pd.merge(transaction,identity,on='TransactionID',how='left')
df = df.columns.str.replace('-','_',regex=True)
```

```python
model = mlflow.sklearn.load_model("runs:/24bc8c1ddc4242af82e89863547beb0d/imb_XGBoost")

prediction = model.predict_proba(df)[:, 1]
predictions = pd.DataFrame({
    'TransactionID': df['TransactionID'],
    'isFraud': prediction
})
predictions.to_csv('/kaggle/working/predictions.csv', index=False)

```
